```python
# Boston AirBnB Reviews
filepath = 'Resources/boston_airbb/reviews.csv'
bos_rev_df = pd.read_csv(filepath)
bos_rev_df.head()
```

```python
bos_cal_df = bos_cal_df.dropna()
```

```python
bos_cal_df.info()
```

```python
# Merged Boston Reviews & Calendar
bos_df = bos_rev_df.merge(bos_cal_df, how='left', on='listing_id')
bos_df.head()
```

```python
bos_df.info()
```

```python
bos_df.shape
```

```python
bos_df = bos_df.drop(columns="date_y")
bos_df = bos_df.rename(columns={'date_x': 'date'})
bos_df
```

```python
bos_df['listing_id'].unique()
```

```python
bos_df.groupby('listing_id').count()
```

```python
bos_df = bos_df.drop_duplicates(subset="listing_id")
bos_df
```

```python
bos_df = bos_df.dropna()
bos_df
```

```python
from datetime import datetime
```

```python
bos_df['date'] = pd.to_datetime(bos_df['date'], format="%Y-%m-%d")
```

```python
bos_df.info()
```

```python
import pymongo
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
db = client.airbb_db
```

```python
collection = db.city
```

```python
bos_df.shape
```

```python

listing_id = list(bos_df.listing_id)
bos_id = list(bos_df.id)
date = list(bos_df.date)
reviewer_id = list(bos_df.reviewer_id)
reviewer_name = list(bos_df.reviewer_name)
comments = list(bos_df.comments)
available = list(bos_df.available)
price = list(bos_df.price)

final = []

for i in range(len(listing_id)):
    post = {
        'listing_id': listing_id[i],
        'id': bos_id[i],
        'date': date[i],
        'reviewer_id': reviewer_id[i],
        'reviewer_name': reviewer_name[i],
        'comments': comments[i],
        'available': available[i],
        'price': price[i]
    }

    final.append(post)

collection.insert_many(final)

    
```
